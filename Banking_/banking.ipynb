```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("Banking.csv")
df.head(5)
```

```python
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Med', 'High']

df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)

df[ 'Income Band'].value_counts().plot()
```

```python
all_cols = [
    'BRId', 'GenderId', 'IAId', 'Amount of Credit Cards', 
    'Nationality', 'Occupation', 'Fee Structure', 'Loyality Classification', 
    'Properties Owned', 'Risk Weighting', 'Income Band'
]

# Keep only columns that exist in df
categorical_cols = [col for col in all_cols if col in df.columns]

for col in categorical_cols:
    print(f"value count for '{col}':")
    display(df[col].value_counts())
```

```python

import seaborn as sns

all_cols = [
    'BRId', 'GenderId', 'IAId', 'Amount of Credit Cards', 
    'Nationality', 'Occupation', 'Fee Structure', 'Loyality Classification', 
    'Properties Owned', 'Risk Weighting', 'Income Band'
]

# Keep only columns that exist in df
categorical_cols = [col for col in all_cols if col in df.columns]

# Choose a column for hue if you want, otherwise set hue=None
hue_col = 'Income Band' if 'Income Band' in df.columns else None

for i, predictor in enumerate(categorical_cols):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue=hue_col)
    plt.title(f"Value counts for {predictor}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for col in categorical_cols:
    if col == "Occupation":
        continue
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col])  
    plt.title(f'Histogram of {col} count')  
    plt.xlabel(col) 
    plt.ylabel("Count")  
    plt.show()
```

```python
numerical_cols = [
    'Estimated Income', 'Superannuation Savings', 'Credit Card Balance',
    'Bank Loans', 'Bank Deposits', 'Checking Accounts',
    'Saving Accounts', 'Foreign Currency Account']

plt.figure(figsize=(15, 20))  # larger height for 8 plots

for i, col in enumerate(numerical_cols):
    plt.subplot(8, 1, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python

numerical_cols = [
    'Estimated Income', 'Superannuation Savings', 'Credit Card Balance',
    'Bank Loans', 'Bank Deposits', 'Checking Accounts',
    'Saving Accounts', 'Foreign Currency Account'
]

plt.figure(figsize=(10, 6))
corr = df[numerical_cols].corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5, square=True)
plt.title("Correlation Heatmap of Numerical Features", fontsize=14)
plt.show()
```
